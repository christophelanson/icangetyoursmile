```python
path_to_images = "../../raw_data/Kaggle_Mask_Lite_dataset/smile"
images_list = sorted(os.listdir(path_to_images))

path_to_resized_images = "../../raw_data/sns/128x128/smile"
resize_size = (128,128)
nb_images = len(images_list)
images_list
```

```python
for image in images_list:
    image_path = f'{path_to_images}/{str(image)}'
    image_HD = Image.open(image_path)
    image_LD = image_HD.resize(resize_size)
    save_path = f'{path_to_resized_images}/{image}'
    image_LD.save(save_path)
    image_LD.format
```
